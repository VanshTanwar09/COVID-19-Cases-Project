```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Group by WHO Region using correct column name and renamed column names if needed
region_summary = df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()

# Plot
plt.figure(figsize=(12,6))
region_summary.set_index('WHO Region').plot(kind='bar', stacked=True)
plt.title("COVID-19 Summary by WHO Region")
plt.ylabel("Number of Cases")
plt.xlabel("WHO Region")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['New cases'], bins=10, kde=True, color='orange')
plt.title("Distribution of New COVID-19 Cases")
plt.xlabel("New Cases")
plt.ylabel("Frequency")
plt.show()
```
