```python
# Avoid division by zero
df['Case Fatality Rate'] = df['Deaths'] / df['Confirmed']
df
```

```python
df['Recovery Rate'] = df['Recovered'] / df['Confirmed']
df.tail()
```

```python
df['Active Rate'] = df['Active'] / df['Confirmed']
df.tail()
```

```python
df['Daily Recovery Rate'] = df['New recovered'] / df['New cases']
df.tail()
```

```python
df['Death Growth Rate'] = df['New deaths'] / df['Deaths']
df.tail()
```

```python
df['Case Growth Rate'] = df['New cases'] / df['Confirmed']
df.head()
```

## MATPLOTLIB DASHBOARD

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("final_dataset.csv")

# Optional: Convert Date column (if time-based plots are used)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Start the dashboard layout
plt.figure(figsize=(20, 15))
plt.suptitle("COVID-19 Dashboard - Cleaned Dataset", fontsize=20)

# 1. Top 10 countries by confirmed cases
plt.subplot(3, 3, 1)
top10 = df.groupby("Country/Region")["Confirmed"].max().sort_values(ascending=False).head(10)
top10.plot(kind="bar", color="skyblue")
plt.title("Top 10 Countries by Confirmed Cases")
plt.ylabel("Confirmed")

# 2. Top 10 countries by deaths
plt.subplot(3, 3, 2)
top10_deaths = df.groupby("Country/Region")["Deaths"].max().sort_values(ascending=False).head(10)
top10_deaths.plot(kind="bar", color="salmon")
plt.title("Top 10 Countries by Deaths")
plt.ylabel("Deaths")

# 3. Time series: Total confirmed cases over time
plt.subplot(3, 3, 3)
if "Date" in df.columns:
    df_grouped = df.groupby("Date")["Confirmed"].sum()
    df_grouped.plot()
    plt.title("Total Confirmed Over Time")
    plt.ylabel("Cases")
    plt.xlabel("Date")

# 4. Pie chart: WHO region distribution
plt.subplot(3, 3, 4)
df["WHO Region"].value_counts().plot.pie(autopct="%1.1f%%", startangle=140)
plt.title("Distribution by WHO Region")
plt.ylabel("")

# 5. Histogram: Distribution of new cases
plt.subplot(3, 3, 5)
df["New cases"].hist(bins=30, color="lightgreen", edgecolor="black")
plt.title("Distribution of New Cases")
plt.xlabel("New Cases")
plt.ylabel("Frequency")

# 6. Heatmap: Correlation between key metrics
plt.subplot(3, 3, 6)
heatmap_cols = ['Confirmed', 'Deaths', 'Recovered', 'Active']
sns.heatmap(df[heatmap_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")

# 7. Scatter plot: New cases vs Active
plt.subplot(3, 3, 7)
sns.scatterplot(data=df, x="New cases", y="Active", alpha=0.3)
plt.title("New Cases vs Active Cases")

# 8. Box plot: Recovery Rate by WHO Region
plt.subplot(3, 3, 8)
sns.boxplot(x="WHO Region", y="Recovery Rate", data=df)
plt.title("Recovery Rate by WHO Region")
plt.xticks(rotation=45)

# 9. Bar chart: Region-wise total confirmed cases
plt.subplot(3, 3, 9)
region_confirmed = df.groupby("WHO Region")["Confirmed"].sum().sort_values(ascending=False)
region_confirmed.plot(kind='bar', color='teal')
plt.title("Region-wise Confirmed Cases")
plt.ylabel("Confirmed Cases")
plt.xlabel("WHO Region")
plt.xticks(rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
